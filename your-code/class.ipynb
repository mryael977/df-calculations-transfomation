```python
import pandas as pd
import numpy as np

vehicles =pd.read_csv('../../lab-import-export/your-code/vehicles/vehicles.csv')
vehicles.head()

vehicles.groupby(['Transmission'])
```

```python
#Here we aggregate 3 different columns and compute their mean based on different transmision values
vehicles.groupby(['Transmission'])['Highway MPG', 'City MPG', 'Combined MPG'].mean()

#In these example we aggregate based on two columns and compute the median CO2 Emission for all combinations
#of fuel type and cylinders

vehicles.groupby(['Fuel Type', 'Cylinders'])['CO2 Emission Grams/Mile'].median()

#Here we produce the mean, median and standard deviation for combined MPG grouped by fuel type

vehicles.groupby(['Fuel Type'])['Combined MPG'].agg(['mean', 'median', 'std'])

```

```python
# Creating aggregate function
from scipy import stats

def agg_mode(x):
    return(stats.mode(x)[0])

vehicles.groupby('Transmission')['Vehicle Class'].agg(agg_mode)
```
