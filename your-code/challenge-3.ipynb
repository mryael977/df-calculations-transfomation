```python
customers_df = orders.groupby(['CustomerID'])['amount_spent'].sum()
customers_df = pd.DataFrame(customers_df)
customers_df
```

```python
customers_df.quantile([.75, .95])
```

```python
prefered_cutomers = customers_df[(customers_df['amount_spent']  >=1661) & (customers_df['amount_spent']< 5840.182)]
prefered_cutomers
```

```python
vip_customers = customers_df[customers_df['amount_spent'] > 5830.182]
vip_customers
```

```python
vip_countries = vip_customers.merge(orders[['CustomerID', 'Country']], on = 'CustomerID', how = 'left')
s = vip_countries['Country'].value_counts()
s
```

```python
preferred_countries = prefered_cutomers.merge(orders[['CustomerID', 'Country']], on = 'CustomerID', how = 'left')
g = preferred_countries['Country'].value_counts()
g
```

```python
a = s+g
a.sort_values(ascending=False)
```
